```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_features(y, sr, title="Audio Features"):
    plt.figure(figsize=(15, 5))

    # Waveform
    plt.subplot(3, 1, 1)
    librosa.display.waveshow(y, sr=sr)
    plt.title(f'Waveform of {title}')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')

    # Spectrogram
    plt.subplot(3, 1, 2)
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Spectrogram')

    # Chromagram
    plt.subplot(3, 1, 3)
    chromagram = librosa.feature.chroma_stft(y=y, sr=sr)
    librosa.display.specshow(chromagram, sr=sr, x_axis='time', y_axis='chroma')
    plt.colorbar()
    plt.title('Chromagram')

    plt.tight_layout()
    plt.show()
```

```python
def visualize_genre_features(base_path, genres, files_per_genre=1):
    for genre in genres:
        print(f"Visualizing {genre}...")
        genre_count = 0

        # Walk through the genre directory, including subdirectories
        for subdir, dirs, files in os.walk(os.path.join(base_path, genre)):
            for file in files:
                if file.endswith('.wav') and genre_count < files_per_genre:
                    file_path = os.path.join(subdir, file)
                    y, sr = librosa.load(file_path, sr=None)
                    plot_features(y, sr, title=f"{genre} - {file}")
                    genre_count += 1

                if genre_count >= files_per_genre:
                    break

            if genre_count >= files_per_genre:
                break
```

```python
base_path = "D:/MusicGenreClassification/ismir04_genre/audio/training"
genres = ["classical", "jazz", "pop", "rock", "punk","electronic","metal","world"]  # List your genres here
visualize_genre_features(base_path, genres, files_per_genre=1)
```
